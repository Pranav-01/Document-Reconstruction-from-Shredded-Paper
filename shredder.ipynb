```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread("p01.png")
image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
print(image.shape)
plt.imshow(image, cmap='gray')
```

```python
shreds = []
for i in range(0, image.shape[1], 100):
    shred = []
    for j in range(0, image.shape[0]):
        shred.append(image[j][i:i+100])
    # print(shred.shape)
    shreds.append(shred)

shreds = np.array(shreds)
print(shreds.shape)
plt.imshow(shreds[0], cmap='gray')
```

```python
for i in range(9):
    cv.imwrite(f"shredded/shred{i}.png", shreds[i])
```
